# Augmentation_Flow_from_dir()

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataGen= ImageDataGenerator(rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    brightness_range=[0.2,2.0],
    shear_range=0.2,
    zoom_range=0.2,
    fill_mode='nearest',
    horizontal_flip=True)
```

## Augmentation

```python
i=0

for batch in dataGen.flow_from_directory('D:\\DATA Science\\DEEP LEARNING\\Augmentation\\vechicles',batch_size=1):
    
    image=batch[0].astype('uint8')
    
    # converting the 4 Dimension image to 3 Dimension
    
    image=np.squeeze(image)
    plt.imshow(image)
    plt.show()
    
    i+=1
    if i>20:
        break
```

## Save the file to directory

```python
i=0

for batch in dataGen.flow_from_directory('D:\\DATA Science\\DEEP LEARNING\\Augmentation\\vechicles',
                                         batch_size=1,
                                         save_to_dir='D:\\DATA Science\\DEEP LEARNING\\Augmentation\\vechicles_aug_images',
                                        save_format='jpeg'):
    
    i+=1
    if i>50:
        break


```
