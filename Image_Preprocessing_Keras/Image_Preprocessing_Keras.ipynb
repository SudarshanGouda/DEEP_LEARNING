# Save Image

```python
from tensorflow.keras.preprocessing.image import save_img
```

```python
dog_image= load_img('dog.png',color_mode='grayscale')
```

```python
plt.imshow(dog_image)
```

```python
DOG_IMAGE= img_to_array(dog_image)
```

```python
save_img('DOG_IMG.jpg',DOG_IMAGE)
```
